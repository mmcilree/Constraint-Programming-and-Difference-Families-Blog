```python
import sympy as sp
```

```python
# Copied from https://stackoverflow.com/questions/40190849/efficient-finding-primitive-roots-modulo-n-using-python
from math import gcd as bltin_gcd

def primRoots(modulo):
    required_set = {num for num in range(1, modulo) if bltin_gcd(num, modulo) }
    return [g for g in range(1, modulo) if required_set == {pow(g, powers, modulo)
            for powers in range(1, modulo)}]
```

```python
# Run through pairs of odd numbers up to 100
for u in range(1, 20)[::2]:
    for l in range(1, 20)[::2]:
        
        factors = sp.factorint(2*u*l + 1)
        
        # Power of a prime of the form 2ul + 1
        if len(factors.keys()) == 1 \
            and list(factors.values())[0] == 1:
            q = list(factors.keys())[0]
            print("q = " + str(list(factors.keys())[0])\
                  + "^" + str(list(factors.values())[0])\
                  + " = 2*" + str(u) + "*"\
                  + str(l) + " + 1")
            print("q = " + str(q))
            print("primitive element = " + str(primRoots(q)))
            
            
```

```python
def print_ext_diffs(sets, modulus):
    count = 0
    results = {}
    for i in range(0, len(sets)):
        for j in range(0, len(sets)):
            if i != j:
                for el1 in sets[i]:
                    for el2 in sets[j]:
                        count += 1
                        print(" & " + str(el1) + " - " + str(el2) + " = " + str((el1 - el2) % modulus), end="")
                        results[str(el1) + " - " + str(el2) + " = "] = (el1 - el2) % modulus
                        if count % 6 == 0:
                            print("\\\\")
    results = dict(sorted(results.items(), key=lambda item: item[1]))
    count = 0
    for diff in results.keys():
        print("& " + diff + str(results[diff]))
        count += 1
        if count % 6 == 0:
            print("\\\\")
```

```python
print_ext_diffs([[0,3,1],[4, 3, 2]], 5)
```

```python
print_ext_diffs([[0,1],[0,1]], 2)
```

```python
def print_tuple_ext_diffs(sets, modulus):
    count = 0
    results = {}
    for i in range(0, len(sets)):
        for j in range(0, len(sets)):
            if i != j:
                for el1 in sets[i]:
                    for el2 in sets[j]:
                        count += 1
                        print(" & " + str(el1) + " - " + str(el2) + " = (" + str((el1[0] - el2[0]) % modulus) + "," + str((el1[1] - el2[1]) % modulus) + ")", end="")
                        results[str(el1) + " - " + str(el2) + " = "] = ((el1[0] - el2[0]) % modulus, (el1[1] - el2[1]) % modulus)
                        if count % 3 == 0:
                            print("\\\\")
    results = dict(sorted(results.items(), key=lambda item: item[1]))
    print("--------")
    count = 0
    for diff in results.keys():
        print("& " + diff + str(results[diff]))
        count += 1
        if count % 3 == 0:
            print("\\\\")
```

```python
print_tuple_ext_diffs([[(1, 0),(0, 1),(2,0),(0,2)],[(1, 1),(1,2), (2, 1), (2, 2)]], 3)
```
